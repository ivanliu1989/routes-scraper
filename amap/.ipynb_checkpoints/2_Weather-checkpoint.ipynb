```python
from pandas.io.json import json_normalize #package for flattening json in pandas df
import pandas as pd

res = requests.get(url_lives)
json_data_lives = json.loads(res.text)
lives = json_data_lives['lives']


res = requests.get(url_forecast)
json_data_forecast = json.loads(res.text)
forecasts = json_data_forecast['forecasts']    
```
