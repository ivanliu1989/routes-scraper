```python
from pandas.io.json import json_normalize #package for flattening json in pandas df
import pandas as pd

for i in range(len(urls)):
    url = urls[i]
    res = requests.get(url)
    json_data = json.loads(res.text)
    count = json_data['count']
    print("Rows: {}".format(count))
    if int(count) == 0:
        break
    pois = json_data['pois']
    new_dat = json_normalize(pois)
    
    if i == 0:
        all_dat = new_dat
    else:
        dats = [all_dat, new_dat]
        all_dat = pd.concat(dats)
    
#     writer.writerow(new_dat)
#     for poi in pois:
#         poi_id = poi['id']
#         parent = poi['parent']
#         name = poi['name']
#         typ = poi['type']
#         typ1 = typ.split(';')[0]
#         typ2 = typ.split(';')[1]
#         typ3 = typ.split(';')[2]
#         business_type = poi['biz_type']
#         address = poi['address']
#         location = poi['location']
#         tel = poi['tel']
#         postcode = poi['postcode']
#         website = pot['website']
#         citycode = pot['citycode']
#         cityname = pot['cityname']
#         adcode = pot['adcode']
#         adname = pot['adname']
#         lon = location.split(',')[0]
#         lat = location.split(',')[1]
#         print(name,typ1,typ2,typ3,lon,lat)
#         writer.writerow([name,typ1,typ2,typ3,lon,lat])
    time.sleep(1)
```

### POI Analysis

```python
import pandas as pd
```

```python
dat = pd.read_csv("./1.csv")
dat.head()
```

```python
from pyecharts import Pie

ty1_class = dat.groupby(['typ1']).agg(['count'])
attr = list(ty1_class.index)
v1 = list(ty1_class['name']['count'])
pie = Pie('POI分布情况')
pie.add("", attr, v1, is_label_show=True, radius=[30, 75])
pie
```

```python
from pyecharts import Bar

ty2_class = dat.groupby(['typ2']).agg(['count'])
bar = Bar('小类POI排行')
attr = list(ty2_class.index)
v2 = list(ty2_class['name']['count'])
bar.add("", attr, v2,xaxis_interval=0,xaxis_rotate=20,xaxis_margin=8,is_label_show=True)
bar
```
