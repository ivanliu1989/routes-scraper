```python
import requests
import json
import time
import csv
```

```python
k = open("../amap_key.txt", "r") 
amap_key = k.read()
rectangle = "116.351147,39.966309;116.357134,39.968727"
extensions = "all"

location = "116.3057764,39.98641364"
radius = 1500

name = "北环大道"
adcode = 440300

url_rec = 'https://restapi.amap.com/v3/traffic/status/rectangle?rectangle={}&key={}&extensions={}'.format(rectangle, amap_key, extensions)
url_circle = 'https://restapi.amap.com/v3/traffic/status/circle?location={}&key={}&radius={}&extensions={}'.format(location, amap_key, radius, extensions)
url_route = 'https://restapi.amap.com/v3/traffic/status/road?name={}&key={}&adcode={}&extensions={}'.format(name, amap_key, adcode, extensions)
```

```python
from pandas.io.json import json_normalize #package for flattening json in pandas df
import pandas as pd

res_rec = requests.get(url_rec)
json_data_rec = json.loads(res_rec.text)

res_circle = requests.get(url_circle)
json_data_circle = json.loads(res_circle.text)

res_route = requests.get(url_route)
json_data_route = json.loads(res_route.text)
```
