```python
import requests
import json
import time
import csv
```

```python
k = open("../amap_key.txt", "r") 
amap_key = k.read()

poi_code = 110202 #[110000, 110200, 110201, 110202, 110203, 110208]
# 110000	风景名胜	风景名胜相关	旅游景点
# 110200	风景名胜	风景名胜	风景名胜
# 110201	风景名胜	风景名胜	世界遗产
# 110202	风景名胜	风景名胜	国家级景点
# 110203	风景名胜	风景名胜	省级景点
# 110208	风景名胜	风景名胜	海滩

ad_code = 310000
# 上海市	310000	021
# 上海市市辖区	310100	021
# 黄浦区	310101	021
# 徐汇区	310104	021
# 长宁区	310105	021
# 静安区	310106	021
# 普陀区	310107	021
# 虹口区	310109	021
# 杨浦区	310110	021
# 闵行区	310112	021
# 宝山区	310113	021
# 嘉定区	310114	021
# 浦东新区	310115	021
# 金山区	310116	021
# 松江区	310117	021
# 青浦区	310118	021
# 奉贤区	310120	021
# 崇明区	310151	021

f_name = '1_ad{}_poi{}.csv'.format(ad_code, poi_code)
f = open(f_name,'a+',encoding='utf-8',newline='')
writer = csv.writer(f)
writer.writerow(['poi_id','parent','name','typ','typ1','typ2','typ3','business_type','address','lon','lat','tel','postcode',
             'website','pcode','pname','citycode','cityname','adcode','adname','entr_location','exit_location',
             'alias','business_area','parking_type','tag','business_area','rating','cost'])

urls = ['https://restapi.amap.com/v3/place/text?types=110000&city={}&offset=20&page={}&key={}&extensions=all'.format(ad_code, str(i), amap_key) for i in range(1,1000)]
```

```python
from pandas.io.json import json_normalize #package for flattening json in pandas df
import pandas as pd

for url in urls:
    res = requests.get(url)
    json_data = json.loads(res.text)
    count = json_data['count']
    print("Rows: {}".format(count))
    if int(count) == 0:
        break
    pois = json_data['pois']
    
    for poi in pois:
        poi_id = poi['id']
        parent = poi['parent']
        name = poi['name']
        typ = poi['type']
        typ1 = typ.split(';')[0]
        typ2 = typ.split(';')[1]
        typ3 = typ.split(';')[2]
        business_type = poi['biz_type']
        address = poi['address']
        location = poi['location']
        lon = location.split(',')[0]
        lat = location.split(',')[1]
        tel = poi['tel']
        postcode = poi['postcode']
        website = poi['website']
        pcode = poi['pcode']
        pname = poi['pname']
        citycode = poi['citycode']
        cityname = poi['cityname']
        adcode = poi['adcode']
        adname = poi['adname']
        entr_location = poi['entr_location']
        exit_location = poi['exit_location']
        alias = poi['alias']
        business_area = poi['business_area']
        parking_type = poi['parking_type'] if 'parking_type' in poi else ''
        tag = poi['tag']
        business_area = poi['business_area']
        rating = poi['biz_ext']['rating']
        cost = poi['biz_ext']['cost']
#         photos_titile = poi['photos']['title']
#         photos_url = poi['photos']['url']
        
        print(poi_id,parent,name,typ,typ1,typ2,typ3,business_type,address,lon,lat,tel,postcode,
             website,pcode,pname,citycode,cityname,adcode,adname,entr_location,exit_location,
             alias,business_area,parking_type,tag,business_area,rating,cost)
        writer.writerow([poi_id,parent,name,typ,typ1,typ2,typ3,business_type,address,lon,lat,tel,postcode,
             website,pcode,pname,citycode,cityname,adcode,adname,entr_location,exit_location,
             alias,business_area,parking_type,tag,business_area,rating,cost
                        ])
    time.sleep(1)
    
f.close()
```
